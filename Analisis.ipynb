```python
import pandas as pd
```

```python
df1 = pd.read_csv('AnimeList1.csv')
df2 = pd.read_csv('AnimeList2.csv')
df3 = pd.read_csv('AnimeList3.csv')
df4 = pd.read_csv('AnimeList4.csv')
df5 = pd.read_csv('AnimeList5.csv')
df6 = pd.read_csv('AnimeList6.csv')
df7 = pd.read_csv('AnimeList7.csv')
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7]).reset_index(drop=True)
df['Name'] = df['Name'].str.replace(',','')
df
```

**DATA CLEANSING**

```python
dfsize = df.groupby('Name')[['Name']].count().rename(columns = {'Name':'Played_anime'}).sort_values('Played_anime')
dfsize
```

```python
# prompt: Using dataframe dfsize: convert to csv

dfsize.to_csv('dfsize.csv')
```

```python
discard = ['yanaginagi', 'Utada Hikaru', 'Togashi Yoshihiro', 'Kanno Yoko', 'Oda Eiichiro',
          'Obata Takeshi', 'Eve', 'Lee Stan', 'Kubo Tite', 'Kon Satoshi', 'Kawahara Reki',
          'Ikuhara Kunihiko', 'Kamijou Ikuru', 'Miyazaki Hayao', 'Suzuki Konomi',
          'Tomino Yoshiyuki', 'Sorachi Hideaki', 'Colquhoun Connor', 'Ado', 'Anno Hideaki',
          'Shinkai Makoto', 'Jackson Samuel L.', 'Hatsune Miku', 'Bosch Johnny Yong', 'Mignogna Vic',
          'Blum Steven', 'Tatum J. Michael', 'Haberkorn Todd', 'Sabat Christopher',
          'Freeman Crispin', 'Papenbrook Bryce', 'Mercer Matthew', 'Bezerra Wendel',
          'Karbowski Brittney', 'Bailey Laura', 'Leigh Cherami', 'Rial Monica', 'Ayres Greg',
          'Lowenthal Yuri', 'Briggs Guilherme', 'Palencia Brina', 'Sinclair Ian', "O'Brien Liam",
          'Schemmel Sean', 'Christian Luci', 'Clinkenbeard Colleen', 'Willingham Travis',
          'Grelle, Jessie James', 'Baker Troy', 'Marchi Jamie', 'Valenzuela Cristina']

df_edit = df[~df.Name.isin(discard)]
df_edit
```

```python
dfMean = df_edit.groupby(['Name', 'CharName','AniName']) \
  .mean() \
  .sort_values('Favorite', ascending=False)

dfMean
```

```python
dfMean = df_edit.groupby(['Name', 'CharName','AniName']) \
  .mean() \
  .groupby('Name') \
  .mean()\
  .sort_values('Favorite', ascending=False)

dfMean
```

```python
played_anime = dfsize['Played_anime']
dfnew = dfMean.join(played_anime)
dfnew = dfnew.reset_index()
dffix = dfnew.drop(['Ranked','Popularity'], axis = 1)

dffix
```

```python
dffix.to_csv('DataSeiyublmkali.csv', index=False)
```

```python
factors = {
    'Rating' : 10000,
    'FavChar' : 10,
    'Members' : 0.1,
    'Played_anime': 100
}

for column, factor in factors.items():
    dffix[column] = dffix[column] * factor

dffix
```

```python
dffix.to_csv('DataSeiyu3coba.csv', index=False)
```
